```python
import numpy as np

import random

#creacion de tablero

def tablero(tamaño):
    return np.full((tamaño,tamaño),"#")

#creacion flota

def crear_barco_CPU(eslora):
    eslora_base = 0
    hori_vert = np.random.randint(0,2,size =1)
    fila = np.random.randint(0,10-eslora, size= (1))
    columna = np.random.randint(0,10-eslora, size= (1))
    tablero_CPU_oculto[fila,columna] = "O"
    while eslora_base != eslora:
        if hori_vert == 0:
            tablero_CPU_oculto[fila+eslora_base,columna] = "O"
            eslora_base = eslora_base +1
            
        else:
            tablero_CPU_oculto[fila,columna+eslora_base] = "O"
            eslora_base = eslora_base +1
   
def crear_barco_jugador(eslora):
    eslora_base = 0
    hori_vert = np.random.randint(0,2,size =1)
    fila = np.random.randint(0,10-eslora, size= (1))
    columna = np.random.randint(0,10-eslora, size= (1))
    tablero_jugador_oculto[fila,columna] = "O"
    while eslora_base != eslora:
        if hori_vert == 0:
            tablero_jugador_oculto[fila+eslora_base,columna] = "O"
            eslora_base = eslora_base +1
            
        else:
            tablero_jugador_oculto[fila,columna+eslora_base] = "O"
            eslora_base = eslora_base +1

#funcion disparar

def disparar_CPU(fila,columna):
    if tablero_jugador_oculto[fila,columna] == "O":
        tablero_jugador[fila,columna] = "O"
        print("Tocado")
        return "Tocado"
    elif tablero_jugador_oculto[fila,columna] == "#" :
        tablero_jugador[fila,columna] = "A"
        print("Agua")
        return "Agua"
    else: 
        print("Error")
        return "Error"

def disparar_jugador(fila,columna):
    if tablero_CPU_oculto[fila,columna] == "O":
        tablero_CPU[fila,columna] = "O"
        print("Tocado")
        return "Tocado"
    elif tablero_CPU_oculto[fila,columna] == "#" :
        tablero_CPU[fila,columna] = "A"
        print("Agua")
        return "Agua"
    else: 
        print("Error")
        return "Error"

```
